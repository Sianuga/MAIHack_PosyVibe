```python
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write

model = MusicGen.get_pretrained('facebook/musicgen-medium')  # or 'melody', 'small', 'stereo'

model.set_generation_params(duration=15)  # seconds
descriptions = ["relaxing ambient synth with soft pads and a slow tempo"]

wav = model.generate(descriptions)  # returns list of torch tensors

# Save to file
audio_write("relaxing_synth", wav[0].cpu(), model.sample_rate, format="wav")
```
